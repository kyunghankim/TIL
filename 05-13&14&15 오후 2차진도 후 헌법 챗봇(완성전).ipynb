```python
#코드 전체의 예로 [45, 439, 648, 102, 4, 649, 650]일 경우
def genData(doc, maxLen, vocabSize):
    for sent in doc: #ex: sent의 길이가 7
        inputs=[]
        targets=[]
        for i in range(1, len(sent)): # range는 1~6까지 반복
            inputs.append(sent[0:i]) #<- [[45], [45, 439]...]
            targets.append(sent[i])  #<- [[439],[638]...]
        y = np_utils.to_categorical(targets,vocabSize) #[439]=>[0,..,0,1,...0]
        inputSeq = sequence.pad_sequences(inputs, maxlen=maxLen)
        yield (inputSeq,y)
        #print("inputs:",inputs,"\n","="*50)
        #print("targets:",targets)   
#    return 
```

```python
# yield와 return의 차이
# 함수 구현시 return하면 그냥 끝나는데
# yield를 하면 객체를 줌
```

```python
for i, (x,y) in enumerate(genData(doc,maxLen, vocabSize)):
    print(i)
    print("x's shape:",x.shape,"\n",x)
    print("y's shape:",y.shape,"\n",y)
```

```python
xdata=[]
ydata=[]
for x,y in genData(doc, maxLen, vocabSize):
    xdata.append(x)
    ydata.append(y)
xdata=np.concatenate(xdata)
ydata=np.concatenate(ydata)
```

```python
import re
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
from konlpy.tag import Okt
from keras import models,layers,optimizers,metrics,preprocessing
```

```python
#태그 단어
PAD='<PADDING>' #패딩
STA='<START>' #시작
END='<END>' #끝
OOV='<OOV>' #out of vocabulary

PAD_index=0
STA_index=1
END_index=2
OOV_index=3

ENCODER_INPUT = 0
DECODER_INPUT = 1
DECODER_TARGET = 2

#한 문장에서 단어 시퀀스의 최대 개수
maxSequences = 30
#임베딩 벡터 차원
embeddingDim = 100
#LSTM 출력 차원
lstmHiddenDim =128
#정규표현식 필터
RE_FILTER = re.compile("[.,!?₩';:~()]")
```

```python
#형태소 분석기
def posTag(sentences):
    tagger=Okt()
    sentencePos=[]
    for sentence in sentences:
        #특수문자 제거
        sentence = re.sub(RE_FILTER,"",sentence) #<- sentence에 있는 RE_FILTER에 해당되면 제거해라
        sentence = " ".join(tagger.morphs(sentence))#<-형태소로 나뉘는 애를 sentence에
        sentencePos.append(sentence)
        
    return sentencePos
```

```python
def convertTextToIndex(sentences, voc, mytype):
    sentencesIndex=[]
    for sentence in sentences:
        sentenceIndex=[]
        if mytype==DECODER_INPUT:
            
            sentenceIndex.extend([voc[STA]]) #<- 1추가
            
        for word in sentence.split():
            
            if voc.get(word) is not None:
                sentenceIndex.extend([voc[word]]) #<- 단어에 해당하는 index가 추가
                
            else: #<-단어가 사전에 없는 경우
                sentenceIndex.extend([voc[OOV]]) #<- OOV를 추가
                
        if mytype==DECODER_TARGET:
            #디코더 출력은 맨 마지막에 end가 추가
            if maxSequences<=len(sentenceIndex):
                sentenceIndex = sentenceIndex[:maxSequences-1]+[voc[END]]
            else:
                sentenceIndex+=[voc[END]]
        else:
            if len( sentenceIndex)> maxSequences:
                sentenceIndex=sentenceIndex[:maxSequences]
        #0으로채우기(paddig)
        sentenceIndex+=[wordToIndex[PAD]]*(maxSequences-len(sentenceIndex))    
        sentencesIndex.append(sentenceIndex)
    return np.asarray(sentencesIndex) #<-list가 array로 바뀌어서 return
```

```python
#모델 생성
#인코더 정의
#입력 문장의 인덱스 sequence를 입력
encoderInputs = layers.Input(shape=(None,))    #<-입력에 대한 정의를 해줌
#Embedding 계층
encoderOutputs = layers.Embedding(len(words), embeddingDim)(encoderInputs)

#<-입력변수설명: 1번째 출력(hidden)shape
encoderOutputs, stateH, stateC = layers.LSTM(lstmHiddenDim,return_state=True,
                                           dropout=0.2,
                                            recurrent_dropout=0.5)(encoderOutputs)
#return_state=True => 상태값이 리턴
#LSTM은 상태가 2개 존재함(셀스테이트,히든스테이트)

encoderStates = [stateH, stateC]

#디코더 정의
#출력 문장의 인덱스 sequence를 입력
decoderInputs = layers.Input(shape=(None,))    #<-출력에 대한 정의를 해줌
#Embedding 계층
decoderEmbedding = layers.Embedding(len(words), embeddingDim)
decoderOutputs = decoderEmbedding(decoderInputs)

#입력변수설명: 1번째 출력(hidden)shape
decoderLSTM= layers.LSTM(lstmHiddenDim,return_state=True,
                                           return_sequences=True, dropout=0.2,
                                            recurrent_dropout=0.5)
decoderOutputs, _,_ = decoderLSTM(decoderOutputs,
                                  initial_state = encoderStates) #<-초기상태:encoderStates
decoderDense = layers.Dense(len(words), activation='softmax')
decoderOutputs = decoderDense(decoderOutputs)
```

```python
model = models.Model( [encoderInputs,decoderInputs],
                     decoderOutputs)
```

```python
model.compile(optimizer='rmsprop',
             loss='categorical_crossentropy',
             metrics=['accuracy'])
#여기까지가 훈련과 관련된 모델 정의
```

```python
#예측 모델에 대한 encoder 정의
encoderModel = models.Model(encoderInputs,
                            encoderStates)

#예측 모델에 대한 decoder 정의
#바로 앞에 있는 decoder의 출력(state)을 입력받아서 예측해야 함
decoderStateInputH = layers.Input(shape=(lstmHiddenDim,))
decoderStateInputC = layers.Input(shape=(lstmHiddenDim,))
decoderStatesInputs = [decoderStateInputH, decoderStateInputC]

#Embedding layer
decoderOutputs = decoderEmbedding(decoderInputs)

#LSTM layer
decoderOutputs, stateH, stateC = decoderLSTM(decoderOutputs,
           initial_state = decoderStatesInputs)
decoderStates = [stateH, stateC]

#Dense layer를 통해 원핫 형식으로 예측단어 인
decoderOutputs = decoderDense(decoderOutputs)

#예측 모델 decoder 설정
models.Model([decoderInputs+decoderStatesInputs,
             [decoderOutputs]+decoderStates])





```

```python
def converIndexToText(indexs,voc):
    #구현
    sentence = ''
    for i in indexs:
        if i==END_INDEX:
            break;
        if voc.get(i) is not None:
            sentence+=voc[i]
        else:
            sentence.extend([voc[OOV_INDEX]])
        sentence+=" "
    return sentence
```

```python
#문장 입력
def makePredictInput(sentence):
    sentences = []
    sentences.append(sentence)
    sentences = posTag(sentences)
    inputSeq = converIndexToText(sentences,
                                 wordToIndex,
                                 ENCODER_INPUT)
    return inputSeq
```

```python
#예측 답변 생성
def generateText(inputSeq):
    #입력을 인코더에 넣고 마지막 상태값을 구하기
    states = encoderModel.predict(inputSeq)
    
    #목표 시퀀스 초기화
    targetSeq = np.zeros((1,1))
    #<START> 시그널을 추가
    targetSeq[0,0] = STA_INDEX #<- <STA>의 tag index
    #index 초기화
    indexs=[]
    
    #decoder 반복
    while 1: #<-계속반복
        decoderOutputs, stateH, stateC = decoderModel.predict( [targetSeq]+states)
        #결과를 원하인코딩형식으로변환
        index= np.argmax(decoderOutputs[0,0,:])
        indexs.append(index)
        #종료 체크
        if index == END_INDEX or len(indexs)>=maxSequences:
            break #<- 최대길이를 넘어가면 while을 빠져나가야함
    
        targetSeq = np.zeros((1,1))
        targetSeq[0,0] = index            #<-STA_INDEX
    
        #decoder의 이전 상태를 다음 decoder예측
        states = [stateH,stateC]
    #index를 문장으로 변환해주면됨
    sentence = converIndexToText(indexs,indexToWord)
    
    return sentence
        
```
