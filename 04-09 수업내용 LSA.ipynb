```python
#원핫인코딩
from konlpy.tag import Okt
```

```python
okt=Okt()
tok=okt.morphs("나는 자연어처리를 학습한다")
#원핫벡터: 단어 집합을 벡터로 표현하는 방식
```

```python
w2i={}
for v in tok:
    if v not in w2i.keys(): # <- key에 존재하지 않는다면
        w2i[v]=len(w2i) # <- w2i의 길이로 index가 부여
print(w2i)
```

```python
#자연어 -> 원핫인코딩 -> 0010000
def ohe(w,w2i):
    ohv=[0]*len(w2i)   #<-0이 7개 들어가있는 리스트
    index=w2i[w]
    ohv[index]=1
    return ohv
print(ohe("자연어",w2i)) # <- '자연어'를 원핫인코딩
```

```python
*알고리즘 공부에서 볼 수 있는 내용:
-run-length 기법: aaaabbbaaaaa => a4b3a5로 압축 ex: BMP파일

-허프만트리(고급)를 이용한 압축, 일반적인 압축은 허프만트리(효율이 좋음)
(검색: 허프만 인코딩): a=>101, b=10, c=1101 ...
-BPE (압축)알고리즘=> 단어 분리에 응용
```

```python
#AAABDAAABAC
#BPE 압축
#연속적인 글자 쌍(2글자)을 구성했을 때 가장 많이 등장
#시행1) AA가 가장많이 등장 => 다른 글자로 치환
#=>소문자 z로 치환
#zABDzABAC

#시행2)  AB가 가장많이 등장 => 다른 글자로 치환
#=>소문자 y로 치환
#zyDzyAC

#시행3)  zy가 가장많이 등장 => 다른 글자로 치환
#=>소문자 x로 치환
#xDxAC
```

```python
*BPE: 단어 분리 알고리즘
    =>글자 단위 -> 단어 집합
ex: train data
    = {low:5, lower:2, newest:6, widest:3}
    (1) 단어 집합(voc):
        => low, lower, newest, widest
    ex: 테스트시 lowest가 입력 -> OOV 문제(모델이 제대로 작동X)
    (2) BPE 알고리즘:
        dict={low:5, lower:2, newest:6, widest:3}
    => 1. 글자 단위로 분리:
         l,o,w,e,r,n,w,s,t,i,d
    => 2. 2개의 순서쌍을 생각할 때, lo가 몇 개? 7개
            we는 몇개? 8개
        순서쌍 중에 가장 많이 등장하는 것 찾기!
        ex는 9개
    *한 글자씩 분리된 것을 'unigram', 2 글자를 한 글자씩 'bigram', 3글자 'trigram'
    => 'n-gram'
    => 3. 가장 빈번하게 등장하는 글자 순서쌍: es
        es를 하나의 글자처럼 취급
        low:5개, lower:2개, new(es)t:6개, wid(es)t:3개
    => 빈도수가 가장 높은 순서쌍을 다시 생각 할 때, lo:7개
        w(es):6개, (es)t:9개
        가장 빈번하게 나온 글자:(es)t
    => 4. low:5, lower:2, new(est):6, wid(est):3개
    => 빈도수가 가장 높은 순서쌍을 또 다시 생각, lo:7개
    => 5. (lo)w:5개, (lo)wer:2개 new(est):6개 wid(est):3개
        계속 반복
    => 10회 반복하면 (논문에서):
        l,o,w,e,r,n,w,s,t,i,d,es,est,lo,low,ne,new,...,wid,widest가 만들어짐
```

```python
*위 BPE알고리즘으로 OOV문제 해결과정:
    1. test과정에서 입력된 "lowest"가 원래는 OOV
    2. 글자 단위 분할 l,o,w,e,s,t
    => (low) / (est)로 나뉨
```

```python
#tfidf 행렬 구성
from sklearn.feature_extraction.text import TfidfVectorizer
vector=TfidfVectorizer(stop_words='english',
               max_features=1000)  #1000개 단어               
```

```python
res=vector.fit_transform(newDf['clean_doc'])
res.shape
```

```python
#SVD-full,truncated
#특이값분해
#행렬 = U*S*VT
#절단된 SVD -> 차원축소의 효과
```

```python
#토픽 숫자:n_components
from sklearn.decomposition import TruncatedSVD
```

```python
svdModel=TruncatedSVD(n_components=20)
```

```python
svdModel.fit(res)
```

```python
import numpy as np
np.shape(svdModel.components_)
# 20개의 토픽과 1000개의 단어
```

```python
#단어 1000개의 이름을 가져오는 함수
terms=vector.get_feature_names() 
terms # <- 중요한 단어들 1000개
```

```python
def getTopic(c,fName,n=10):
    for i,t in enumerate(c): # t: topic
        print("토픽: %d:" % (i+1),
              [(fName[i], 
                t[i].round(5)) for i in t.argsort()[:-n-1:-1]])
getTopic(svdModel.components_,terms)    
```
