```python
#코드 전체의 예로 [45, 439, 648, 102, 4, 649, 650]일 경우
def genData(doc, maxLen, vocabSize):
    for sent in doc: #ex: sent의 길이가 7
        inputs=[]
        targets=[]
        for i in range(1, len(sent)): # range는 1~6까지 반복
            inputs.append(sent[0:i]) #<- [[45], [45, 439]...]
            targets.append(sent[i])  #<- [[439],[638]...]
        y = np_utils.to_categorical(targets,vocabSize) #[439]=>[0,..,0,1,...0]
        inputSeq = sequence.pad_sequences(inputs, maxlen=maxLen)
        yield (inputSeq,y)
        #print("inputs:",inputs,"\n","="*50)
        #print("targets:",targets)   
#    return 
```

```python
# yield와 return의 차이
# 함수 구현시 return하면 그냥 끝나는데
# yield를 하면 객체를 줌
```

```python
for i, (x,y) in enumerate(genData(doc,maxLen, vocabSize)):
    print(i)
    print("x's shape:",x.shape,"\n",x)
    print("y's shape:",y.shape,"\n",y)
```

```python
xdata=[]
ydata=[]
for x,y in genData(doc, maxLen, vocabSize):
    xdata.append(x)
    ydata.append(y)
xdata=np.concatenate(xdata)
ydata=np.concatenate(ydata)
```

```python
#모델만들기
model=Sequential()
model.add(Embedding(vocabSize,100,input_length=maxLen)) #<- 1165,100
#Embed:끼워넣기 퍼가기
model.add(LSTM(100, return_sequences=False))
model.add(Dropout(0.5))
model.add(Dense(vocabSize,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',
             optimizer='rmsprop', #<-RMSprop() 함수로 써도됨
             metrics=['accuracy'])
```

```python
hist=model.fit(xdata,ydata,epochs=500,batch_size=800)
```

```python
import re
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
from konlpy.tag import Okt
from keras import models,layers,optimizers,metrics,preprocessing
```

```python
#태그 단어
PAD='<PADDING>' #패딩
STA='<START>' #시작
END='<END>' #끝
OOV='<OOV>' #out of vocabulary

PAD_index=0
STA_index=1
END_index=2
OOV_index=3

ENCODER_INPUT = 0
DECODER_INPUT = 1
DECODER_TARGET = 2

#한 문장에서 단어 시퀀스의 최대 개수
maxSequences = 30
#임베딩 벡터 차원
embeddingDim = 100
#LSTM 출력 차원
lstmHiddenDim =128
#정규표현식 필터
RE_FILTER = re.compile("[.,!?₩';:~()]")
```

```python
#형태소 분석기
def posTag(sentences):
    tagger=Okt()
    sentencePos=[]
    for sentence in sentences:
        #특수문자 제거
        sentence = re.sub(RE_FILTER,"",sentence) #<- sentence에 있는 RE_FILTER에 해당되면 제거해라
        sentence = " ".join(tagger.morphs(sentence))#<-형태소로 나뉘는 애를 sentence에
        sentencePos.append(sentence)
        
    return sentencePos
```

```python
""" shift + tab : 내어쓰기,
    tab: 들여쓰기
    """
```

```python
def convertTextToIndex(sentences, voc, mytype):
    sentencesIndex=[]
    for sentence in sentences:
        sentenceIndex=[]
        if mytype==DECODER_INPUT:
            sentenceIndex.extend([voc[STA]]) #<- 1추가
        for word in sentence.split():
            if voc.get(word) is not None:
                sentenceIndex.extend([voc[word]]) #<- 단어에 해당하는 index가 추가
            else: #<-단어가 사전에 없는 경우
                sentenceIndex.extend([voc[OOV]]) #<- OOV를 추가
                
        if mytype==DECODER_TARGET:
            #디코더 출력은 맨 마지막에 end가 추가
            if maxSequences<=len(sentenceIndex):
                sentenceIndex = sentenceIndex[:maxSequences-1]+[voc[END]]
            else:
                sentenceIndex+=[voc[END]]
        else:
            if len( sentenceIndex)> maxSequences:
                sentenceIndex=sentenceIndex[:maxSequences]
        #0으로채우기(paddig)
        sentenceIndex+=[wordToIndex[PAD]]*(maxSequences-len(sentenceIndex))    
        sentencesIndex.append(sentenceIndex)
    return np.asarray(sentencesIndex) #<-list가 array로 바뀌어서 return
```
